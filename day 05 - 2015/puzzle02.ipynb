```python
def read_file(path):
    with open(path, "r") as readfile:
        a = readfile.read()
        lines = a.split('\n')
    return lines
```

```python
def has_repeated_char_letter_in_between(input):
    for i in range(len(input) - 1):  # Loop through the list up to the second-to-last element
        if input[i] == input[i - 2]:  # Check if the current element equals the next one
            return True
    return False
```

```python
def two_pairs(input):
    for index, char in enumerate(input):
        if index > 0:
            a = input[index-1]
            b = input[index]
            c = a + b
            if input.count(c) > 1:
                return True
    return False
```

```python
n = 0

input = read_file('input01.txt')
for line in input:
    if has_repeated_char_letter_in_between(line) and two_pairs(line):
        n += 1
        print(line, 'nice')
    else:
        print(line, 'naughty')
```
